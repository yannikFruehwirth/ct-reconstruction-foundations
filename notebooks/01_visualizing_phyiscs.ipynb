# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import iradon
from physics.ct_ops import CTProjector
```

# Understand Filtering in Backprojection

```python
projector = CTProjector(size=256)
theta = np.linspace(0., 180., 180, endpoint=False)
sinogram = projector.project(theta)

unfiltered = iradon(sinogram, theta=theta, filter_name=None)

filtered = iradon(sinogram, theta=theta, filter_name='ramp')

# Plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.imshow(unfiltered, cmap='gray')
ax1.set_title("Simple Backprojection\n(Blurry / 1/r artifact)")
ax2.imshow(filtered, cmap='gray')
ax2.set_title("Filtered Backprojection\n(Sharp / Ramp Filter)")
plt.show()
```
